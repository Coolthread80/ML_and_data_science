```python
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Input, Flatten, Dense, Conv2D
from keras.models import Model,Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from keras.datasets import fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
NUM_CLASSES = 10
CLASSES = np.array(["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"])
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train = to_categorical(y_train, NUM_CLASSES)
y_test = to_categorical(y_test, NUM_CLASSES)
```

```python
img_train_classes = CLASSES[np.argmax(y_train, axis = -1)]
```

```python
fig, ax = plt.subplots(3, 3)
for i in range(3):
    for j in range(3):
        k = np.random.randint(0, len(y_train))
        ax[i][j].imshow(x_train[k], cmap="gray")
        ax[i][j].set_title(img_train_classes[k])
        ax[i][j].axis('off')
plt.show()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28), name="Input_layer"))
model.add(Dense(128, activation='relu', name="Hidden_layer_1"))
model.add(Dense(128, activation='relu', name="Hidden_layer_2"))
model.add(Dense(128, activation='relu', name="Hidden_layer_3"))
model.add(Dense(128, activation='relu', name="Hidden_layer_4"))
model.add(Dense(128, activation='relu', name="Hidden_layer_5"))
model.add(Dense(128, activation='relu', name="Hidden_layer_6"))
model.add(Dense(128, activation='relu', name="Hidden_layer_7"))
model.add(Dense(128, activation='relu', name="Hidden_layer_8"))
model.add(Dense(128, activation='relu', name="Hidden_layer_9"))
model.add(Dense(128, activation='relu', name="Hidden_layer_10"))
model.add(Dense(128, activation='relu', name="Hidden_layer_11"))
model.add(Dense(128, activation='relu', name="Hidden_layer_12"))
model.add(Dense(128, activation='relu', name="Hidden_layer_13"))
model.add(Dense(128, activation='relu', name="Hidden_layer_14"))
model.add(Dense(128, activation='relu', name="Hidden_layer_15"))
model.add(Dense(128, activation='relu', name="Hidden_layer_16"))
model.add(Dense(128, activation='relu', name="Hidden_layer_17"))
model.add(Dense(128, activation='relu', name="Hidden_layer_18"))
model.add(Dense(10, activation='softmax', name="Output_layer"))
```

```python
model.summary()
```

```python
opt = Adam(learning_rate=0.0005)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
model.fit(x_train
          , y_train
          , batch_size=32
          , epochs=50
          , shuffle=True)
```

```python
model.evaluate(x_test, y_test)
```

```python
preds = model.predict(x_test)
pred = CLASSES[np.argmax(preds,axis=-1)]
act = CLASSES[np.argmax(y_test, axis = -1)]
```

```python
fig, ax = plt.subplots(3, 4)

fig.subplots_adjust(hspace=0.6, wspace=0.6)

for i in range(3):
    for j in range(4):
      k = np.random.randint(0, len(y_test))
      ax[i][j].imshow(x_test[k], cmap="gray")
      ax[i][j].set_title(f"Pred = {pred[k]}" + '\n' + f"Real = {act[k]}")
      ax[i][j].axis('off')
plt.show()
```
